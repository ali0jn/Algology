```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('algology_dataset.xls').iloc[:863, :]
df = df.rename(columns = {'Diagnosis':'Diagnosis_1', 'R/L':'R/L_1', 'Procedure':'Procedure_1'}) 

df['Pain Level - After 1 hour'] = df['Pain Level - After 1 hour'].replace('?','')

df.insert(8, 'Diagnosis_2', df['Diagnosis_1'])
for i in range(len(df['Diagnosis_1'])):
    if type(df['Diagnosis_1'].iloc[i]) == float: continue
    splitted = df['Diagnosis_1'].iloc[i].split('+')
    splitted_2 = df['Diagnosis_1'].iloc[i].split('/')
    if len(splitted) == 2:
        df['Diagnosis_1'].iloc[i] = splitted[0]
        df['Diagnosis_2'].iloc[i] = splitted[1]
    elif len(splitted_2) == 2:
        df['Diagnosis_1'].iloc[i] = splitted_2[0]
        df['Diagnosis_2'].iloc[i] = splitted_2[1]

df['Diagnosis_1'] = df['Diagnosis_1'].str.upper()
df['Diagnosis_2'] = df['Diagnosis_2'].str.upper()
df['Diagnosis_1'] = df['Diagnosis_1'].str.strip()
df['Diagnosis_2'] = df['Diagnosis_2'].str.strip()
df['Diagnosis_1'] = df['Diagnosis_1'].replace('[.]', '', regex=True)
df['Diagnosis_2'] = df['Diagnosis_2'].replace('[.]', '', regex=True)
df['Diagnosis_1'] = df['Diagnosis_1'].replace('ADEZİV KAPSÜLİT', 'ADHEZIV KAPSÜLIT')
df['Diagnosis_2'] = df['Diagnosis_2'].replace('ADEZİV KAPSÜLİT', 'ADHEZIV KAPSÜLIT')
df['Diagnosis_1'] = df['Diagnosis_1'].replace('KOKSİDİNİ', 'KOKSIDINI')
df['Diagnosis_2'] = df['Diagnosis_2'].replace('KOKSİDİNİ', 'KOKSIDINI')
df['Diagnosis_1'] = df['Diagnosis_1'].replace('KOKSARTROZ', 'KOKSARTOZ')
df['Diagnosis_2'] = df['Diagnosis_2'].replace('KOKSARTROZ', 'KOKSARTOZ')
df['Diagnosis_1'] = df['Diagnosis_1'].replace('KOKSİGODİNİ', 'KOKSIGODINI')
df['Diagnosis_2'] = df['Diagnosis_2'].replace('KOKSİGODİNİ', 'KOKSIGODINI')




df.insert(10, 'R/L_2', df['R/L_1'])
for i in range(len(df['R/L_1'])):
    if type(df['R/L_1'].iloc[i]) == float: continue
    splitted = df['R/L_1'].iloc[i].split('+')
    if len(splitted) == 2:
        df['R/L_1'].iloc[i] = splitted[0]
        df['R/L_2'].iloc[i] = splitted[1]
    if len(splitted) == 3:
        df['R/L_1'].iloc[i] = splitted[0]
        df['R/L_2'].iloc[i] = splitted[1]

df['R/L_1'] = df['R/L_1'].str.upper()
df['R/L_2'] = df['R/L_2'].str.upper()
df['R/L_1'] = df['R/L_1'].str.strip()
df['R/L_2'] = df['R/L_2'].str.strip()
df['R/L_1'] = df['R/L_1'].replace('[.:]', '', regex=True)
df['R/L_2'] = df['R/L_2'].replace('[.:]', '', regex=True)
df['R/L_1'] = df['R/L_1'].replace(('ORT HAT','ORTHAT','ORT'), 'ORTA HAT')
df['R/L_2'] = df['R/L_2'].replace(('ORT HAT','ORTHAT','ORT'), 'ORTA HAT')

df.insert(12, 'Procedure_2', df['Procedure_1'])
for i in range(len(df['Procedure_1'])):
    if type(df['Procedure_1'].iloc[i]) == float: continue
    splitted = df['Procedure_1'].iloc[i].split('+')
    splitted_2 = df['Procedure_1'].iloc[i].split('/')
    if len(splitted)>=2:
        df['Procedure_1'].iloc[i] = splitted[0]
        df['Procedure_2'].iloc[i] = splitted[1]
    if len(splitted_2)>=2:
        df['Procedure_1'].iloc[i] = splitted_2[0]
        df['Procedure_2'].iloc[i] = splitted_2[1]

df['Procedure_1'] = df['Procedure_1'].str.upper()
df['Procedure_2'] = df['Procedure_2'].str.upper()
df['Procedure_1'] = df['Procedure_1'].str.strip()
df['Procedure_2'] = df['Procedure_2'].str.strip()
df['Procedure_1'] = df['Procedure_1'].replace('[.]', '', regex=True)
df['Procedure_2'] = df['Procedure_2'].replace('[.]', '', regex=True)
df['Procedure_1'] = df['Procedure_1'].replace('TFESI', 'TFESE')
df['Procedure_2'] = df['Procedure_2'].replace('TFESI', 'TFESE')
df['Procedure_1'] = df['Procedure_1'].replace('İLESİ', 'ILESI')
df['Procedure_2'] = df['Procedure_2'].replace('İLESİ', 'ILESI')
df['Procedure_1'] = df['Procedure_1'].replace('FMDRF', 'FEMDRF')
df['Procedure_2'] = df['Procedure_2'].replace('FMDRF', 'FEMDRF')
df['Procedure_1'] = df['Procedure_1'].replace('GANGİM', 'GANGIM')
df['Procedure_2'] = df['Procedure_2'].replace('GANGİM', 'GANGIM')
df['Procedure_1'] = df['Procedure_1'].replace('OMUZ İA', 'OMUZ IA')
df['Procedure_2'] = df['Procedure_2'].replace('OMUZ İA', 'OMUZ IA')
df['Procedure_1'] = df['Procedure_1'].replace('FASET İA', 'FASET IA')
df['Procedure_2'] = df['Procedure_2'].replace('FASET İA', 'FASET IA')
df['Procedure_1'] = df['Procedure_1'].replace('OMUZ İA ENJ', 'OMUZ IA ENJ')
df['Procedure_2'] = df['Procedure_2'].replace('OMUZ İA ENJ', 'OMUZ IA ENJ')
df['Procedure_1'] = df['Procedure_1'].replace(('GENIKÜLER RF', 'GENIKULAT RF', 'GENIKULAR RF'), 'GENİKÜLER RF')
df['Procedure_2'] = df['Procedure_2'].replace(('GENIKÜLER RF', 'GENIKULAT RF', 'GENIKULAR RF'), 'GENİKÜLER RF')
df['Procedure_1'] = df['Procedure_1'].replace(('SPLANNIK BLOK', 'SPLANKNIK BLOK'), 'SPLANIK BLOK')
df['Procedure_2'] = df['Procedure_2'].replace(('SPLANNIK BLOK', 'SPLANKNIK BLOK'), 'SPLANIK BLOK')
df['Procedure_1'] = df['Procedure_1'].replace('GANG İMPAR BLOKAJI', 'GANGIMPAR BLOKAJI')
df['Procedure_2'] = df['Procedure_2'].replace('GANG İMPAR BLOKAJI', 'GANGIMPAR BLOKAJI')
df['Procedure_1'] = df['Procedure_1'].replace('SUPERIOR HIPOGASTRIK BL', 'SÜPERIOR HIPOGASTRIK BLOK')
df['Procedure_2'] = df['Procedure_2'].replace('SUPERIOR HIPOGASTRIK BL', 'SÜPERIOR HIPOGASTRIK BLOK')
df['Procedure_1'] = df['Procedure_1'].replace('GANG IMPAR', 'GANGİMPAR')
df['Procedure_2'] = df['Procedure_2'].replace('GANG IMPAR', 'GANGİMPAR')
df['Procedure_1'] = df['Procedure_1'].replace('STELLAT GANG BLOKAJI', 'SATELLAT GANG BLOKAJI')
df['Procedure_2'] = df['Procedure_2'].replace('STELLAT GANG BLOKAJI', 'SATELLAT GANG BLOKAJI')
df['Time'] = df['Time'].replace('[^0-9]', '', regex=True)
df['Height'] = df['Height'].replace('[/]','.', regex=True)
df['Height'] = df['Height'].replace(1957, np.nan)
df['Radiation Dose'] = df['Radiation Dose'].replace('0,7484-0,14039','0.444395')
df['Radiation Dose'] = df['Radiation Dose'].replace(',','.')

for i in range(len(df['Age'])):
    if len(str(df['Age'].iloc[i])) == 6:
        df['Age'].iloc[i] = 2018 - df['Age'].iloc[i]
# df.to_excel(r'F:\Machine Learning\Projects\Algology\dataset(column duplication app, version.alireza).xls')
```
